## opencv Image Read

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image = cv2.imread('/content/photo.jpg')

# Calculate the quarter height and quarter width of the image
height, width = image.shape[:2]
quarter_height, quarter_width = height / 2, width / 2

# Define the transformation matrix T for an affine transformation
T = np.float32([[1, 0, quarter_width], [0, 1, quarter_height]])

# Apply the affine transformation to get the transformed image
img_trans = cv2.warpAffine(image, T, (width, height))



# Create a figure and add three subplots for the original, transformed, and resized images
fig, axes = plt.subplots(1, 2, figsize=(15, 5))

# Display the original image
axes[0].axis('off')
axes[0].imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
axes[0].set_title('Original Image')

# Display the transformed image
axes[1].imshow(cv2.cvtColor(img_trans, cv2.COLOR_BGR2RGB))
axes[1].set_title('Transformed Image')
axes[1].axis('off')



plt.show()
```
